```python
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
paths = glob.glob("../results/*.npz")

attn_dict = np.load(paths[-2], allow_pickle=True)
print(paths[-2])
attn = attn_dict["visual_attention_weight_sums"]
```

```python
mean_attn = attn.mean(axis=0)
std_attn = attn.std(axis=0)

plt.figure(figsize=(12, 4))
plt.bar(range(len(mean_attn)), mean_attn, yerr=std_attn, capsize=5)
plt.xlabel("Layer Index")
plt.ylabel("Attention Overlap with Vision Tokens")
plt.title("Visual Attention Overlap with Vision Tokens: Hugging Face LLaVA")
plt.xticks(range(len(mean_attn)))
plt.ylim(0, 0.2)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
